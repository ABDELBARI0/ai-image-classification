```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

import PIL
from PIL import Image

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers,models
from tensorflow.keras.models import Sequential


import cv2
import os
import glob
import random
```

```python
general_path = "/content/drive/MyDrive/PetImages"
categories = ["Cat","Dog"]

images = []
labels = []

width,hight = 150 , 150

for category in categories :

    path = os.path.join(general_path,category)

    for anemales in os.listdir(path) :

      anemale_path = os.path.join(path,anemales)

      img = cv2.imread(anemale_path)

      if img is not None:

            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            img=cv2.resize(img ,(width,hight))

            if category == "Cat" :

                  images.append([img,0])
                  labels.append(category)

            else :

                  images.append([img,1])
                  labels.append(category)

      else:
            pass

encoded_label = LabelEncoder().fit_transform(labels)
```

```python
fueater = []
leabels = []

for f,l in images :
    fueater.append(f)
    leabels.append(l)
```

```python
fueater = np.array(fueater)
leabels = np.array(leabels)
```

```python
x_train , x_test , y_train , y_test  = train_test_split(fueater,leabels,train_size=0.81)
```

```python
# y_train # 20247
# y_test # 4750

y_train = np.array(y_train)
y_test = np.array(y_test)

print("Shape of x_train:", x_train.shape)
print("Shape of y_train:", y_train.shape)
print("Data type of x_train:", x_train.dtype)
print("Data type of y_train:", y_train.dtype)
print("Min value in x_train:", x_train.min())
print("Max value in x_train:", x_train.max())
```

```python
width,hight = 150 , 150
```

```python
model = Sequential([

    layers.Conv2D(16,(3,3), padding='same', activation='relu', input_shape=(width,hight,1)),
    layers.MaxPool2D(pool_size=(2, 2),padding='same'),
    layers.Conv2D(32,(3,3), padding='same', activation='relu'),
    layers.MaxPool2D(pool_size=(2, 2),padding='same'),
    layers.Conv2D(64,(3,3), padding='same', activation='relu'),
    layers.MaxPool2D(pool_size=(2, 2),padding='same'),
    # layers.Conv2D(64,(3,3), padding='same', activation='relu'),
    # layers.MaxPool2D(pool_size=(2, 2),strides=(1,1),padding='same'),
    layers.Flatten(),
    layers.Dense(128,activation="sigmoid"),
    layers.Dense(64,activation="sigmoid"),
    layers.Dense(32,activation="sigmoid"),
    layers.Dense(2,activation="softmax")

])
```

```python
model.compile(optimizer ="adam",loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,
                         validation_data=(x_test,y_test),
                         epochs=10,
                         batch_size=100,
                         verbose=1,
                         callbacks=[tf.keras.callbacks.EarlyStopping(
                                            patience=10,
                                            monitor='val_accuracy',#"val_loss",
                                            restore_best_weights=True)])
```

```python
model.evaluate(
    x=x_test,
    y=y_test,
    batch_size=100)
```
